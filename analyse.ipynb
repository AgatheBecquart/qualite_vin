# ANALYSE DES DIFFÉRENTS CRITÈRES QUI INFLUENCENT LA QUALITÉ DU VIN

```python
import pandas as pd

df = pd.read_csv("winequality-red.csv")

```

```python
df.columns = ['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar', 'chlorides',
                'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'quality']
```

## La corrélation

```python
corr = df.corr()
corr
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6) )
sns.heatmap(corr, cmap="coolwarm", annot=True)
plt.title("Matrice de Corrélations")
plt.show()
```

En analysant la corrélation, on peut déjà souligner que l'alcool influence beaucoup sur la qualité du vin. Les sulfates et l'acide citrique ont également une part d'influence. 

# Distribution du dataset

```python
sns.displot(df['quality'])
plt.title('Distribution des vins selon leur qualité')
plt.xlabel('Qualité')
plt.ylabel('Nombre de vin')
```

On remarque que le dataset contient une grande majorité de vins classés en 5 et 6 pour leur qualité. On prévoit déjà un algorithme plus performant pour les données de ce type. 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Créer un boxplot pour chaque variable selon la qualité
for var in df.columns[:-1]:
    fig, ax = plt.subplots(figsize=(8,6))
    ax.set_title(f'Boxplot de {var} selon la qualité')
    ax.set_xlabel('Qualité')
    ax.set_ylabel(var)
    sns.boxplot(x='quality', y=var, data=df, ax=ax)
    plt.show()
```

Avec ces graphiques il est bien confirmé que l'alcool, les sulfates ainsi que l'acide citrique ont une influence positive sur la qualité. 
Tandis que le Ph, la densité, les chlorides ainsi que l'acidité volatiles ont une influence négative sur le qualité.
